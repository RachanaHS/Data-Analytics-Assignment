```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
#from pandas.tools.plotting import scatter_matrix
import numpy as np
import seaborn as sns
import statistics as st
from scipy.stats import norm
import re
df=pd.read_csv("IndiaVotes_PC__All_States_2014.csv")

df.Party.value_counts().plot(kind="pie")
plt.legend(bbox_to_anchor=(2,0.7),loc='best')
plt.show()

```

```python
g = sns.FacetGrid(df)  
g.map(sns.barplot, 'Type',"Electors", ci = 95)  
plt.show()
```

```python
g = sns.FacetGrid(df)  
g.map(sns.barplot, 'Type',"Votes", ci = 95)  
plt.show()
```

```python

sns.factorplot(x="Type", y="Votes", hue="State", data=df,kind="bar", ci = None)
plt.show()
```

```python
df.Party.value_counts().plot(kind = "bar")
plt.xlabel('Party')
plt.ylabel('Number of places won')
plt.title('Bar Plot')
plt.show()
```
