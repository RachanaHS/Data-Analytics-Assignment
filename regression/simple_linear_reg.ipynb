```python
%matplotlib inline

# Imports
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
import sklearn
import numpy as np

df = pd.read_csv('iris.csv')
```

```python
snsdata = df.drop(['petal_width','sepal_width'], axis=1)
```

```python
sns.regplot(x='petal_length', y='sepal_length', data=df, fit_reg=False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = np.asarray(train_test_split(df['petal_length'], df['sepal_length'], test_size=0.1))
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train.values.reshape(-1,1), y_train.values.reshape(-1,1))
```

```python
print('Score: ', reg.score(X_test.values.reshape(-1,1), y_test.values.reshape(-1,1)))
```

```python
x_line = np.arange(5,25).reshape(-1,1)
sns.regplot(x=df['petal_length'], y=df['sepal_length'], data=df, fit_reg=False)
plt.plot(x_line, reg.predict(x_line))
plt.show()
```
